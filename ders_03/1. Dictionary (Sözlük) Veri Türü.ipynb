# DICTIONARY (dict) Veri Türü
**key**<b style="color:red">:</b>**value** (anahtar<b style="color:red">:</b>değer) şeklinde ikili yapıdan oluşan bir veri türüdür. Programlamada çok sık kullanılan JSON yapısına benzemektedir. Bazı programlama dillerinde Map veri türü olarak da geçmektedir. Bu veri türünü tam anlamıyla bir sözlük gibi de düşünebiliriz. Aşağıda bulunan tablo, Python'daki **dict** veri türü ile pratik bir şekilde ifade edilebilir.

Sözlük (**dict**) veri türü, süslü parantezle başlar ve süslü parantezle biter. Her bir **anahtar:değer** (key:value) ikilisi, alt alta yazılabildiği gibi, tüm bir **dict** verisi, tek satırda da yazılabilir. Aşağıdaki örnekte, 6 adet **key:value** ikilisinden oluşan bir sözlük örneği verilmiştir. Kısaca aşağıdaki sözlük yapısı, 6 elemandan oluşmuştur ve her bir eleman **key:value** şeklinde ikili bir yapıya sahiptir diyebiliriz.

```python
sozluk = {
    "Python":"Bir programlama dilidir.",
    "Internet":"En büyük bilgisayar ağıdır.",
    "Su":"Bir içecektir.",
    "Güneş":"Isı ve ışık kaynağıdır.",
    "Makarna":"Bir yemek çeşididir.",
    "Gmail":"Ücretsiz e-posta hizmeti verir."
}
 
sozluk
```

```python
sozluk["Python"]
```

# Boş bir sözlük (dict) oluşturmak

```python
sozluk1 = dict()  
sozluk2 = {}      # list olunca [], tuple olunca (), dict olunca da {} olur.

print(type(sozluk1), type(sozluk2))
```

# Dolu bir sözlük oluşturmak
Sözlüğü oluşturmak için, istersek süslü parantezler yerine, **dict()** fonksiyonunu da kullanabiliriz.<br>
**dict()** fonksiyonu  içerisinde, anahtar-değer ikililerini <b style="color:red">:</b> karakteri yerine, <b style="color:red">=</b> karakteri ile yazmaya özen göstermeliyiz!

```python
sozluk1 = dict(anahtar="değer", toplam=3)     # dict fonksiyonu içerisinde     key=value
sozluk2 = {"anahtar": "değer", "toplam": 3}   # süslü parantezler içerisinde   key:value

print(sozluk1, sozluk2)
```

# Liste yapısı içinde Sözlük oluşturmak

Aşağıda bir excel tablosuna ait veriler yer almaktadır.<br>
Bunun için **list** içerisinde **dict** veri türünü kullanabiliriz. Bu kullanıma ait örnekler aşağıda verilmiştir.

```python
# list içerisinde dict oluşturmak
tablo = [
    {"Adı":"Ahmet", "Soyadı":"Öztürk", "Konum":"İstanbul", "Meslek":"Doktor", "Yaş":32},
    {"Adı":"Mustafa", "Soyadı":"Yalçın", "Konum":"Ankara", "Meslek":"Mühendis", "Yaş":29},
    {"Adı":"Ayşe", "Soyadı":"Aydın", "Konum":"İzmir", "Meslek":"Öğretmen", "Yaş":28},
    {"Adı":"Merve", "Soyadı":"Kaya", "Konum":"İstanbul", "Meslek":"Öğretmen", "Yaş":27},
    {"Adı":"Hasan", "Soyadı":"Taner", "Konum":"İzmir", "Meslek":"Teknisyen", "Yaş":30},
    {"Adı":"Hüseyin", "Soyadı":"Candaş", "Konum":"Konya", "Meslek":"Öğretmen", "Yaş":26},
    {"Adı":"Sema", "Soyadı":"Pehlivan", "Konum":"İstanbul", "Meslek":"Fotoğrafçı", "Yaş":25}  
]
```

```python
type(tablo)
```

```python
type(tablo[2])
```

```python
tablo
```

Örneğin, tablodaki 3. kişinin tüm bilgilerini almak isteyelim.

```python
tablo[2]
```

Şimdi de tablodaki 3. kişinin meslek bilgisini alalım.

```python
tablo[2]["Meslek"]
```

**Not:** Liste içinde sözlük oluşturabildiğimiz gibi, Tuple içinde de sözlük oluşturabiliriz.

# Sözlük yapısı içerisinde Liste oluşturmak

Aşağıda başka bir tabloya ait veriler yer almaktadır.<br>
Bunun için **dict** içerisinde **list** veri türünü kullanabiliriz. Bu kullanıma ait örnekler aşağıda verilmiştir.

```python
# dict içerisinde list oluşturmak
sozluk = {
    "Öğrenciler": ["Ali Aydın", "Berna Öztürk", "Elif Demir"],
    "Bölüm": ["Matematik", "Fizik", "Bilgisayar Mühendisliği"],
    "Şehirler": ["İstanbul", "Ankara", "İzmir"],
    "Yatay Geçiş": [2021, 2022, 2022],
    "GANO": [3.14, 3.5, 3.66]
}
```

```python
type(sozluk)
```

```python
type(sozluk["Öğrenciler"])
```

```python
sozluk
```

Örneğin, tablodaki tüm öğrencilerin bilgilerini almak isteyelim.

```python
sozluk["Öğrenciler"]
```

Şimdi de tablodaki öğrencilerden, üçüncü sırada yatay geçiş yapan öğrencinin **şehir** bilgisini edinelim.

```python
sozluk["Şehirler"][2]
```

**Not:** Sözlük içinde liste oluşturabildiğimiz gibi, sözlük içinde tuple veri türünü de aynı mantıkla oluşturabiliriz.

# Sözlük yapısında indisler
Sözlük (**dict**) veri türünde, daha önce işlediğimiz; **str**, **list** ve **tuple** veri türlerindeki gibi, otomatik bir indis numaralandırma mantığı yoktur.<br>
İndis mantığı **key** (anahtar) adı verilen yapılarla sağlanır. 

```python
araba = {
    "model": 2022,
    "marka": "Audi",
    "renk":  "Kırmızı"
}

araba
```

```python
print(araba[0]) # hata meydana gelir. çünkü numaralandırılmış bir indisleme yoktur!
```

```python
araba["model"]
```

Sözlük (**dict**) veri türünde; **str**, **list** veya **tuple**'da kullandığımız numaralı indisleme mantığını aşağıdaki gibi kurabiliriz.

```python
otomobil = {
    1: "audi",
    2: "beyaz",
    3: 2023
}

otomobil
```

```python
otomobil[1]  # hata meydana gelmez. çünkü kullanılan indis aslında bir anahtardır! (key)
```

```python
otomobil[0]  # hata meydana gelir. çünkü 0 adlı bir anahtar (key), sözlük içinde tanımlı değildir!
```

# Sözlük anahtarlarına ait değerlerin değiştirilmesi

```python
araba = {
    "model": 2020,
    "marka": "Audi",
    "renk":  "Kırmızı"
}

araba
```

```python
araba
```

# Sözlüğe yeni key:value (anahtar:değer) ikilisi eklemek

**str**, **list** ve **tuple** veri türlerinde aşağıdaki gibi, indisler kullanılarak yeni değerler eklenemiyordu. <br>
Fakat **dict** veri türünde, atama ile yeni **anahtar:değer** ikilileri eklemek mümkündür.
